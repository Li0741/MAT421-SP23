**17.1 Interpolation Problem Statement**

Assume we are given a data set of independent data values xi and dependent data vlaues yi, where i = 1, ..., n. 

The goal of interpolation is to find the estimation funtion for the given set of points.

Interpolation does not require the user to have an underlying model for the data, but it require to inform the user about the quality of the interpolation. 

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

x = [1,2,3,4]
y = [1,5,3,4]
plt.figure(figsize = (10,8))
plt.plot(x,y,'bo', 2.5,4,'ro') 
plt.title('Interpolation data point')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**17.2 Linear Interpolation**

In linear interpolation, the estimated point is assumed to lie on the line joining the nearest points to the left and right.

Following example finds the linear interpolation at based on the data x = [0, 1, 3, 5], y = [1, 5, 3, 2.5]. 

```python
from scipy.interpolate import interp1d 
import matplotlib.pyplot as plt 
plt.style.use('seaborn-poster') 

x = [0, 1, 3, 5] 
y = [1, 5, 3, 2.5] 
 
f = interp1d(x, y) 
y_hat = f(2) 
print(y_hat) 
```

```python
plt.figure(figsize = (10,8)) 
plt.plot(x, y, '-ob') 
plt.plot(2, y_hat, 'ro') 
plt.title('Linear Interpolation at x = 2') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show() 
```

**17.3 Cubic Spline**

In cubic spline interpolation, the interpolating function is a set of piecewise cubic functions.

Following is the cubic spline interpolation of the data set x = [0, 2, 4, 6] and y = [1, 7, 6, 4] for 0 <= x <= 7.

```python
from scipy.interpolate import CubicSpline 
import numpy as np 
import matplotlib.pyplot as plt 
 
plt.style.use('seaborn-poster') 

x = [0, 2, 4, 6] 
y = [1, 7, 6, 4] 

f = CubicSpline(x, y, bc_type='natural') 
x_new = np.linspace(0, 7, 100) 
y_new = f(x_new) 

plt.figure(figsize = (10,8)) 
plt.plot(x_new, y_new, 'b') 
plt.plot(x, y, 'ro') 
plt.title('Cubic Spline Interpolation') 
plt.xlabel('x') 
plt.ylabel('y') 
plt.show() 
```
